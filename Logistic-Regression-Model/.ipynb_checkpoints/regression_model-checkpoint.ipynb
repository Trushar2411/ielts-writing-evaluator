```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Ridge, LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data_path = "../data/filter_dataset.csv"
```

# Model

```python
class RegressionModel:
    def __init__(self, base_model, tf_idf_max_features, tf_idf_n_gram_range, data_path):
        self.tf_idf_max_features = tf_idf_max_features
        self.tf_idf_n_gram_range = tf_idf_n_gram_range
        self.base_model = base_model
        self.data_path = data_path
        self.data_preprocessing()
        self.tf_idf_vectorization()
        
    def data_preprocessing(self):
        # Load dataset
        df = pd.read_csv(self.data_path)
        df = df[['Question', 'Essay', 'Overall']].dropna()
        
        # Preprocess: Combine question and answer
        df['Text'] = df['Question'] + " " + df['Essay']
        
        # Split data into features and target
        self.X = df['Text']
        self.y = df['Overall']
        
    def tf_idf_vectorization(self):
        tfidf = TfidfVectorizer(max_features=self.tf_idf_max_features, ngram_range=self.tf_idf_n_gram_range)
        X_tfidf = tfidf.fit_transform(self.X)
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X_tfidf, self.y, test_size=0.2, random_state=42)

    def train(self):    
        self.model = self.base_model
        self.model.fit(self.X_train, self.y_train)

    def evaluate(self): 
        # Predict on test 
        y_pred = self.model.predict(self.X_test)

        # Loss
        mae = mean_absolute_error(self.y_test, y_pred)
        mse = mean_squared_error(self.y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(self.y_test, y_pred)
        
        print(f"MAE:  {mae:.2f}")
        print(f"MSE:  {mse:.2f}")
        print(f"RMSE: {rmse:.2f}")
        print(f"R²:   {r2:.2f}")
        
        # Visualization
        y_true = np.array(self.y_test)
        sns.set(style="whitegrid")
        plt.figure(figsize=(8, 6))
        
        sns.scatterplot(x=y_true, y=y_pred, color='royalblue', s=50, alpha=0.7)
        
        sns.regplot(x=y_true, y=y_pred, scatter=False, color='red', line_kws={"linewidth": 2})
        
        plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='green', linestyle='--', label='Ideal')
        plt.xlabel("Actual Scores")
        plt.ylabel("Predicted Scores")
        plt.title(f"Predicted vs Actual Essay Scores using {type(self.model).__name__}")
        plt.legend()
        plt.tight_layout()
        plt.show()
    
```

# Test on different hyper-parameters

```python
ridge_regression_model = RegressionModel(base_model=Ridge(),  tf_idf_max_features=5000,  tf_idf_n_gram_range=(1,3), data_path=data_path)
ridge_regression_model.train()
ridge_regression_model.evaluate()
```

```python
linear_regression_model = RegressionModel(base_model=LinearRegression(),  tf_idf_max_features=5000,  tf_idf_n_gram_range=(1,3), data_path=data_path)
linear_regression_model.train()
linear_regression_model.evaluate()
```
